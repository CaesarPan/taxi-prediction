```python
import os
import pandas as pd
```

```python
df = pd.read_pickle('./assets/clean-data/activity_master.pickle')
kmeans_df = pd.read_pickle('./assets/clean-data/kmeans-df.pickle')

kmeans_dict = {}
for ids, locationID in zip(kmeans_df["ClusterGlobalIDs"].values, kmeans_df["LocationID"]):
    if ids:
        for clus in ids[0]:
            kmeans_dict[clus] = locationID
```

```python
df["LocationID"] = df["cluster_id"].apply(lambda x: kmeans_dict[x])
```
